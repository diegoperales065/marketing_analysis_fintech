```python

def limpieza_fintech(df_fintech):
    #Cambio de nombre de columnas para mayor claridad
    df_fintech = df_fintech.rename(columns={'job':'job_type','marital':'marital_status','education':'education_level','loan':'has_personal_loan','housing':'has_housing_loan','month':'contact_month','day_of_week':'contact_day','contact':'contact_method','duration':'call_duration','default':'credit_default','campaign':'contact_attempts','pdays':'previously_contacted','previous':'previous_contacts','poutcome':'previous_campaign_outcome','y':'subscribed','euribor3m':'euribor_3m_rate','nr.employed':'total_employment','cons.price.idx':'consumer_price_index','emp.var.rate':'employment_variation_rate','cons.conf.idx':'consumer_confidence_index'})
    # Dividir euribor_3m_rate por 10000
    df_fintech['euribor_3m_rate'] = df_fintech['euribor_3m_rate'] / 10000
    # Dividir consumer_price_index por 1000
    df_fintech['consumer_price_index'] = df_fintech['consumer_price_index'] / 1000
    #Eliminar duplicados considerando todas las columnas excepto 'outcome'
    columnas_duplicados = df_fintech.columns.tolist()[:-1]
    df_fintech = df_fintech.drop_duplicates(subset=columnas_duplicados)
    
    #Reemplazar 'unknown' por NaN en la columna 'credit_default' y mapear 'yes' a 1 y 'no' a 0
    df_fintech["credit_default"] = df_fintech["credit_default"].replace("unknown", np.nan)
    df_fintech["credit_default"] = df_fintech["credit_default"].map({"yes":1, "no":0})
    
    #Eliminar filas con 'unknown' en las columnas de préstamos
    df_fintech = df_fintech[(df_fintech["has_housing_loan"] != "unknown") & (df_fintech["has_personal_loan"] != "unknown")]
    #Mapear 'yes' a 1 y 'no' a 0 en las columnas de préstamos
    df_fintech["has_housing_loan"] = df_fintech["has_housing_loan"].map({"yes": 1, "no": 0})
    df_fintech["has_personal_loan"] = df_fintech["has_personal_loan"].map({"yes": 1, "no": 0})
    
    #Mapear 'yes' a 1 y 'no' a 0 en la columna de suscripción
    df_fintech["subscribed"] = df_fintech["subscribed"].map({"yes": 1, "no": 0})

    # Modificar 'previously_contacted' para que 999 se considere como no contactado (0) y cualquier otro valor como contactado (1)
    df_fintech['previously_contacted'] = df_fintech['previously_contacted'].apply(lambda x: 0 if x == 999 else 1)
    
    # Añadir columna 'is_new_campaign_client' con mapeo
    df_fintech['is_new_campaign_client'] = (df_fintech['previous_contacts'] == 0).map({True: 1, False: 0})
    
    # Create high_contact_attempts column based on condition
    df_fintech['high_contact_attempts'] = (df_fintech['contact_attempts'] > 3).map({False: 0, True: 1})
    return df_fintech



df_fintech_clean = limpieza_fintech(df_fintech.copy())
df_fintech_clean.head()
```
